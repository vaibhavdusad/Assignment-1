# Derivative using First principle 

```python
def derivative(x,y):
    m = np.array([a-b for a,b in zip(x,x[1:])]) 
    n = np.array([a-b for a,b in zip(y,y[1:])])
    derivative = n/m
    return derivative
```

# Numerical solver to solve a^x = derivative(a^x)

```python
def solve():
    x = np.linspace(-3,3,100) #Keeping domain to be [-3,3]
    diff = 10 #Initializing parameters
    a = 1 
    #Solving the eqautions upto precision of 0.01
    while(diff>0.0001):
        a = a + 0.001
        y = np.power(a,x)
        z = derivative(x,y)
        diff = np.sum((y[1:]-z))
    print('Value of a for which both function as well as derivative produces same value is approximately equal to {}'.format(a))
    return 
```
