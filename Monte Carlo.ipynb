```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Generating Random x, y coordinates

```python
x = np.random.rand(10000)
y = np.random.rand(10000)
```

# Calculating euclidean distance of points from center(0,0)

```python
z = np.sqrt(np.power(x,2) + np.power(y,2))
```

Calculating area of quadrant on the basis of probablity of number of points inside the quadrant. 
#(pi*r*r/4)/a*a = probability of number of points inside the quadrant
#a=1,r=1 so pi = 4*probability

```python
m = z[z<=1]
area  = m.shape[0]/10000
pi = area*4
print('Pi : {} , Approximation : {}'.format(np.pi,pi))
```

# Segregating points

```python
x_inside = x[z<1]
x_outside =x[z>=1]
y_inside = y[z<1]
y_outside = y[z>=1]
```

# Plotting the results

```python
plt.figure()
ax = sns.scatterplot(x_outside,y_outside)
ax = sns.scatterplot(x_inside,y_inside)
ax.set(xlabel='X Coordinate', ylabel='Y coordinate',title = 'Monte Carlo Pi Approximation')
plt.show()
```
